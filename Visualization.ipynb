```python
import matplotlib.pyplot as plt
import numpy as np
categories = np.array([0, 2, 1, 1, 1, 2])
colormap = np.array(['r', 'g', 'b'])
fig = plt.figure()
ax = fig.add_subplot(111, title="My Plot")
x = [0, 1, 2, 3, 4, 5]
y = [0, 3, 6, 9, 12, 6]
plt.scatter(x, y, s=[70], c=colormap[categories])
plt.show()
```

<h3>Bar Plot</h3>


```python
fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111)
ax.set(title='Bar Plot',
      xlabel='quarter',
      ylabel='sales')
quarters = [1, 2, 3]
sales_2017 = [25782, 35783, 36133]
ax.bar(quarters, sales_2017,color='red', width=0.6, edgecolor='black')
ax.set_xticks(quarters)
ax.set_xticklabels(['Q1', 'Q2', 'Q3'])
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title='Avg. Quarterly Sales',
      xlabel='Quarter', ylabel='Sales (in millions)')
quarters = [1, 2, 3]
x1_index = [0.8, 1.8, 2.8];
x2_index = [1.2, 2.2, 3.2]
sales_2016 = [28831, 30762, 32178];
sales_2017 = [25782, 35783, 36133]
ax.bar(x1_index, sales_2016, color='lightblue', width=0.4, edgecolor='black', label='2016')
ax.bar(x2_index, sales_2017, color='coral', width=0.4, edgecolor='black', label='2017')
ax.set_xticks(quarters)
ax.set_xticklabels(['Q1', 'Q2', 'Q3'])
ax.legend()
plt.show()
```

<h2>Barh</h3>

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title='Avg. Quarterly Sales',
      xlabel='Sales (in millions)', ylabel='Quarter')
quarters = [1, 2, 3]
sales_2017 = [25782, 35783, 36133]
ax.barh(quarters, sales_2017, height=0.6, color='coral')
ax.set_yticks(quarters)
ax.set_yticklabels(['Q1-2017', 'Q2-2017', 'Q3-2017'])
plt.show()
```

<h3>Pie</h3>

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
ax.set(title='Avg. Quarterly Sales')
sales_2017 = [25782, 35783, 36133]
quarters = ['Q1-2017', 'Q2-2017', 'Q3-2017']
ax.pie(sales_2017, labels=quarters,shadow=True,counterclock=True, startangle=90, autopct='%1.1f%%')
plt.show()
```

<h3>Hist</h3>

```python
import numpy as np
np.random.seed(100)
x = 60 + 10*np.random.randn(1000)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title="Distribution of Student's Percentage",
      ylabel='Proportion', 
        xlabel='Percentage')
ax.hist(x, color='blue', bins=30, density=True)
plt.show()
```

<h3>BoxPlot</h3>


```python
import numpy as np
np.random.seed(100)
x = 60 + 10*np.random.randn(1000)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title="Box plot of Student's Percentage",
      xlabel='Class', ylabel='Percentage')
ax.boxplot(x, labels=['A'], bootstrap=10000)
plt.show()
```

```python
import numpy as np
np.random.seed(100)
x = 50 + 10*np.random.randn(1000)
y = 70 + 25*np.random.randn(1000)
z = 30 + 5*np.random.randn(1000)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title="Box plot of Student's Percentage",
      xlabel='Class', ylabel='Percentage')
ax.boxplot([x, y, z], labels=['A', 'B', 'C'], notch=True, bootstrap=10000)
plt.show()
```

```python
import numpy as np
np.random.seed(100)
x = 50 + 10*np.random.randn(1000)
y = 70 + 25*np.random.randn(1000)
z = 30 + 5*np.random.randn(1000)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title="Box plot of Student's Percentage",
      xlabel='Class', ylabel='Percentage')
bplot1=ax.boxplot([x, y, z], labels=['A', 'B', 'C'], patch_artist=True,notch=True, bootstrap=10000)
plt.show()
colors = ['pink', 'lightblue', 'lightgreen']
for bplot in (bplot1):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)
plt.show()      
```

```python
import numpy as np
np.random.seed(100)
x = 50 + 10*np.random.randn(1000)
y = 70 + 25*np.random.randn(1000)
z = 30 + 5*np.random.randn(1000)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title="Box plot of Student's Percentage",
      xlabel='Class', ylabel='Percentage')
ax.boxplot([x, y, z], labels=['A', 'B', 'C'],vert=False,notch=True, bootstrap=10000)
plt.show() 
```
